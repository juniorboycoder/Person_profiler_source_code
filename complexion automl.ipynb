```python
import tensorflow as tf
from IPython.display import Image, display

import matplotlib.pyplot as plt
import tensorflow_hub as hub

import numpy as np

from deep_autoviml import deep_autoviml as deepauto
```

### 2. SPECIFY THE IMAGE DIRECTORY , AND THE IMAGE HEIGHT , WIDTH AND CHANNELS

```python
image_dir = 'C:/Users/admin/Desktop/1.TOTORIAL/PRACTICE new/projects/vision project/complexion - Copy'
img_height = 224
img_width = 224
img_channels = 10
```

```python
keras_model_type =  "image" ## always try "auto" first, then "basic", "deep", "big deep", "giant deep", "cnn1" etc.
keras_options = {"early_stopping": True, 'lr_scheduler': 'rlr', "epochs": 20}  ### always set early_stopping to True first and then change it to False
model_options = {'tuner':"optuna", "max_trials":10, 'cat_feat_cross_flag':True,
                 'image_directory': image_dir, 'image_height': img_height,
                 'image_width':img_width, 'image_channels':img_channels }
```

```python
model, dicti = deepauto.fit("", "", keras_model_type=keras_model_type,
                            project_name='complexion',
                            save_model_flag=False, model_options=model_options,
                            keras_options=keras_options, use_my_model='')
```

### 3.USE TRAINED  MODEL TO PREDICT ON TEST DATASET

```python
predictions=deepauto.predict_images(image_dir, model, cat_vocab_dict=dicti)
```

```python
test_ds = tf.keras.preprocessing.image_dataset_from_directory('C:/Users/admin/Desktop/1.TOTORIAL/PRACTICE new/projects/vision project/complexion - Copy/test',
  seed=111,
  image_size=(img_height, img_width),
  batch_size=128)
```
